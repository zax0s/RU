```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load(img_file):
    with open(img_file, mode='rb') as file: # b is important -> binary
            TempImage = np.fromfile(file, dtype=np.float32)
    return(np.copy(TempImage).reshape(256,256,128,order='F'))
def load2D(img_file):
    with open(img_file, mode='rb') as file: # b is important -> binary
            TempImage = np.fromfile(file, dtype=np.float32)
    return(np.copy(TempImage).reshape(256,256,1,order='F'))
```

## Make 2D Image and Masks

```python
image_GT = load('zczubal.i')
image_GT[:,:,62].reshape(256*256,order='F').astype('float32').tofile('Zubal_2D.img')
ThalamusVOI=np.copy(image_GT)
ThalamusVOI[ThalamusVOI!=14]=0
ThalamusVOI[ThalamusVOI!=0]=1
ThalamusVOI[:,:,62].reshape(256*256,order='F').astype('float32').tofile('Thalamus_2DVOI.img')
CortexVOI=np.copy(image_GT)
CortexVOI[CortexVOI!=8]=0
CortexVOI[CortexVOI!=0]=1
CortexVOI[:,:,62].reshape(256*256,order='F').astype('float32').tofile('Cortex_2DVOI.img')
```

### Show images

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3,figsize=(14,4))
ax1.imshow(image_GT[:,:,62],cmap='gray',vmin=0)
ax1.set_title('Zubal Phantom',size=16)
ax2.imshow(ThalamusVOI[:,:,62],cmap='gray',vmin=0)
ax2.set_title('Thalamus VOI',size=16)
ax3.imshow(CortexVOI[:,:,62],cmap='gray',vmin=0)
ax3.set_title('Cortex VOI',size=16)
plt.savefig('Input_Images.png',bbox_inches='tight',dpi=400)
```

### Show test cases

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3,figsize=(14,4))
CortexVOI = load2D('Cortex_2DVOI.img')
CortexVOI_eroded = load2D('Cortex_eroded.img')
ax1.imshow(CortexVOI,cmap='gray',vmin=0)
ax1.set_title('Cortex VOI',size=16)
ax2.imshow(CortexVOI_eroded,cmap='gray',vmin=0)
ax2.set_title('Eroded Cortex VOI',size=16)
ax3.imshow(CortexVOI,cmap='gnuplot',interpolation='none')
ax3.imshow(CortexVOI_eroded,cmap='gray', alpha=0.4,interpolation='none')
ax3.set_title('Difference',size=16)
plt.savefig('Cortex_example_k3.png',bbox_inches='tight',dpi=400)
```
