```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load(img_file):
    with open(img_file, mode='rb') as file: # b is important -> binary
            TempImage = np.fromfile(file, dtype=np.float32)
    return(np.copy(TempImage).reshape(256,256,128,order='F'))
```

```python
image_GT = load('zczubal.i')
```

```python
plt.imshow(image_GT[:,:,62],cmap='gray',vmin=0)
plt.title('Original Zubal Phantom',size=16)
image_GT[:,:,62].reshape(256*256,order='F').astype('float32').tofile('Zubal_2D.img')
```

```python
ThalamusVOI=np.copy(image_GT)
ThalamusVOI[ThalamusVOI!=14]=0
ThalamusVOI[ThalamusVOI!=0]=1
plt.imshow(ThalamusVOI[:,:,62],cmap='gray',vmin=0)
plt.title('Thalamus VOI - Zubal Phantom',size=16)
ThalamusVOI[:,:,62].reshape(256*256,order='F').astype('float32').tofile('Thalamus_2DVOI.img')
```

```python
CortexVOI=np.copy(image_GT)
CortexVOI[CortexVOI!=8]=0
CortexVOI[CortexVOI!=0]=1
plt.imshow(CortexVOI[:,:,62],cmap='gray',vmin=0)
plt.title('Cortex VOI - Zubal Phantom',size=16)
CortexVOI[:,:,62].reshape(256*256,order='F').astype('float32').tofile('Cortex_2DVOI.img')
```
